```python
#1-opt
#solutionMap: array which represents assignments (solution)
#cost: cost of the solution
#data: customer locations
#p: number of facilities
def one_opt(solutionMap,cost,p,data):
    candidate=cp(solutionMap)
    br=1
    while(1):
        for i in range(654):
            for j in range(1,p-1):
                new_facility=(solutionMap[i]+j)%p
                new_solution=cp(solutionMap)
                new_solution[i]=new_facility
                new_cost=find_cost(new_solution,data,p)
                if new_cost < cost:
                    cost = new_cost
                    solutionMap = cp(new_solution)
                    br=0
        if br == 1:
            break
        else:
            br = 1
    return solutionMap,cost   
```

```python
#2-opt
#solutionMap: array which represents assignments (solution)
#cost: cost of the solution
#data: customer locations
#p: number of facilities
def two_opt(solutionMap,cost,p,data):
    candidate=cp(solutionMap)
    br=1
    brk = 0
    while(brk < 10):
        for i in range(0,653):
            for j in range(1,p-1):
                new_facility=(solutionMap[i]+j)%p
                new_solution=cp(solutionMap)
                new_solution[i]=new_facility
                for ii in range(i+1,654):
                    for jj in range(1,p-1): 
                        new_facility2=(solutionMap[ii]+jj)%p
                        new_solution[ii]=new_facility2
                        new_cost=find_cost(new_solution,data,p)
                        if new_cost < cost:
                            cost = new_cost
                            solutionMap = cp(new_solution)
                            br=0
                            brk =0 
                        else:
                            brk += 1
                            br = 1
                        if brk > 200:
                                return solutionMap,cost   
        if br == 1:
            break
        else:
            br = 1
    return solutionMap,cost   
```

```python
#3-opt
#solutionMap: array which represents assignments (solution)
#cost: cost of the solution
#data: customer locations
#p: number of facilities
def three_opt(solutionMap,cost,p,data):
    candidate=cp(solutionMap)
    br=1
    brk = 0
    while(brk < 10):
        for i in range(0,652):
            for j in range(1,p-1):
                new_facility=(solutionMap[i]+j)%p
                new_solution=cp(solutionMap)
                new_solution[i]=new_facility
                for ii in range(i+1,653):
                    for jj in range(1,p-1): 
                        new_facility2=(solutionMap[ii]+jj)%p
                        new_solution[ii]=new_facility2
                        for iii in range(ii+1,654):
                            for jjj in range(1,p-1): 
                                new_facility3=(solutionMap[iii]+jjj)%p
                                new_solution[iii]=new_facility3
                                new_cost=find_cost(new_solution,data,p)
                                if new_cost < cost:
                                    cost = new_cost
                                    solutionMap=cp(new_solution)
                                    br=0
                                    brk = 0
                                else:
                                    brk += 1
                                if brk > 200:
                                        return solutionMap,cost   
        if br == 1:
            break
        else:
            br = 1
    return solutionMap,cost   
```

```python
#shaking step
#solutionMap: array which represents assignments (solution)
#cost: cost of the solution
#data: customer locations
#k: neighborhood structure (k-opt)
#p: number of facilities
def shaking(solutionMap,cost,k,data,p):
    if k==1:
        i=np.random.randint(0,654)
        new_facility=(solutionMap[i]+1)%p
        new_solution=cp(solutionMap)
        new_solution[i]=new_facility
        new_cost=find_cost(new_solution,data,p)
    elif k==2:
        i=np.random.randint(0,653)
        ii=i+1
        new_facility=(solutionMap[i]+1)%p
        new_facility2=(solutionMap[ii]+1)%p
        new_solution=cp(solutionMap)
        new_solution[i]=new_facility
        new_solution[ii]=new_facility2
        new_cost=find_cost(new_solution,data,p)
    else:
        i=np.random.randint(0,652)
        ii=i+1
        iii=ii+1
        new_facility=(solutionMap[i]+1)%p
        new_facility2=(solutionMap[ii]+1)%p
        new_facility3=(solutionMap[iii]+1)%p
        new_solution=cp(solutionMap)
        new_solution[i]=new_facility
        new_solution[ii]=new_facility2
        new_solution[iii]=new_facility3
        new_cost=find_cost(new_solution,data,p)
    return new_solution,new_cost
```

```python
#basic VNS algorithm
#kMax: 3 - for changing the assignment of 1, 2 and 3 customers
#iterMax: stopping condition
#p: number of facilities
def basicVNS(kMax,iterMax,p):
    #data: customer locations
    data=load_data()
    #solutionMap: array which represents assignments
    solutionMap=cp(init_facilities(p))
    cost=find_cost(solutionMap,data,p)
    i=1
    brandnew_cost = 0
    for i__ in range(iterMax):
        k=1
        while k<= kMax:  
            new_solutionMap,new_cost=shaking(solutionMap,cost,k,data,p)
            if k==1:
                brandnew_solutionMap,brandnew_cost=one_opt(new_solutionMap,new_cost,p,data)
            elif k==2:
                brandnew_solutionMap,brandnew_cost=two_opt(new_solutionMap,new_cost,p,data)
            else:
                brandnew_solutionMap,brandnew_cost=three_opt(new_solutionMap,new_cost,p,data)
            if(brandnew_cost< cost):
                solutionMap=cp(brandnew_solutionMap)
                cost = brandnew_cost
                k=1
            else:
                k=k+1
    return sqrt_cost(solutionMap,data,p)
        
```
